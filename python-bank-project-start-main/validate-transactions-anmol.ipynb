```python
import great_expectations as gx
import pandas as pd
import warnings
warnings.filterwarnings("ignore", message="`result_format` configured at the Validator-level*")

# Load the data
df = pd.read_csv("./data/transactions.csv")

# Create the ephemeral GX context
context = gx.get_context()

# Add a pandas datasource
data_source = context.data_sources.add_pandas(name="pandas")

# Add a dataframe asset
data_asset = data_source.add_dataframe_asset(name="transactions_data")

# Define the batch (entire DataFrame)
batch_definition = data_asset.add_batch_definition_whole_dataframe(name="batch_def")
batch = batch_definition.get_batch(batch_parameters={"dataframe": df})

# Create the expectation suite with a name
suite = gx.core.expectation_suite.ExpectationSuite(name="transactions_suite")

# Get the validator using the suite
validator = context.get_validator(batch=batch, expectation_suite=suite)

# Add expectations
validator.expect_column_values_to_be_between("amount", min_value=0.01, max_value=100000)

# Validate
results = validator.validate()

# Print results
print(results)
```

```python
import great_expectations as gx
import pandas as pd
import warnings
warnings.filterwarnings("ignore", message="`result_format` configured at the Validator-level*")

# Load the data
df = pd.read_csv("./data/sebank_customers_with_accounts.csv")

# Create the ephemeral GX context
context = gx.get_context()

# Add a pandas datasource
data_source = context.data_sources.add_pandas(name="pandas")

# Add a dataframe asset
data_asset = data_source.add_dataframe_asset(name="customers_with_accounts_data")

# Define the batch (entire DataFrame)
batch_definition = data_asset.add_batch_definition_whole_dataframe(name="batch_def")
batch = batch_definition.get_batch(batch_parameters={"dataframe": df})

# Create the expectation suite with a name
suite = gx.core.expectation_suite.ExpectationSuite(name="customers_suite")

# Get the validator using the suite
validator = context.get_validator(batch=batch, expectation_suite=suite)

# Add expectations

# Add expectations for data quality
validator.expect_column_values_to_not_be_null("Customer")
validator.expect_column_values_to_not_be_null("Address")
validator.expect_column_values_to_not_be_null("Phone")
validator.expect_column_values_to_match_regex("Phone", r"^(\+46|0)\d+$")  # Swedish phone number validation
validator.expect_column_values_to_match_regex("Personnummer", r"^\d{6}-\d{4}$")  # Swedish personal number
validator.expect_column_values_to_match_regex("BankAccount", r"^SE\d{2}[A-Z0-9]{18,30}$")  # Swedish IBAN format  # Swedish IBAN format

# Validate
results = validator.validate()

# Print results
print(results)
```
